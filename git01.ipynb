**git log :** describe los últimos commits 

* cuando te aparece esto, cada commit aparece con un codigo-commit especifico para ese commit

**git checkout :** te lleva a un commit especifico cuando uno pone git checkout commit-code

* cuando uno se devuelve, se puede mezclar con git switch -m ??? creo para empezar otra branch desde ese commit. Sino, se puede devolver al último commit con git switch 

**git diff :** se ve especificamente que se cambio de commit en commit, uno puede poner ej. creo q es 

**git tag :** se marca un commit en especifico para tenerlo a mano 

* uno se devuelve al commit con git checkout y despues lo taggea: 

git checkout commit-code

git tag nombre-tag 

para que despues sea tan simple como: git checkout nombre-tag

**.gitignore :** es un archivo donde podemos usar wildcards para evitar archivos innecesarios

**.gitkeep :** servira para contrarrestar lo anterior, ej, si solo queremos archivos .txt en un directorio /d/ , entonces el .gitignore iria en el general pero dentro de /d/ deberìa estar un .gitkeep para mantener los .txt

**git commit --amend :** we can fix the message on the commit

* if you catch it right away: git commit --amend and you'll be able to fix it right away 

**git restore :** restore the state of a file we edited to the last commit

**git revert :** if we commit a big ass mistake, and we commit it and after some time we realize. we gotta look to the git log and having the commit (have that commit-code)

* you can revert with: git restore commit-code. AAAND we add a new commit that does the reverse. 

**git reset :** if we wanna pretend that the mistakes didn't happen. we use reset. **git reset HEAD~(amount of commits we wanna go back)**

* we only delete the commits, if we wanna erase the changes on the files, we can use **git reset --hard**
* it is a lineal thing.

Bringing changes together 

**git cherry-pick :** will allow us to cherry pick commits into the branch that we are working on 

* if we wanna add the commit commit-code to the main branch, we just gotta **git cherry-pick commit-code** and we just added it as a new commit. 

* it'll be conflicts when there are parallel changes that affect each other (the same part of the same file)

**git merge :** merging changes from another branch. with big ass projects it can cause some conflicts

* because it includes ALL of the changes made in both branches 

* they recommend using other approaches 

**git rebase :** its weird, it takes the trunk (bunch of commits) of the other branch and puts it at the end of the main branch, so the commits become lineal once again. we do git rebase new_branch 

* the info of the new_branch is not lost just far back 

**git merge --squash :** you just merge the branch into a single "squashed" commit and preserves the linear history
